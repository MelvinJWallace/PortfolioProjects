This project will focus on analyzing the wage difference between black, white, and hispanic men and women with different educational levels. The data for this dataset begins in 1973 and ends in 2022. This data was retrieved from the Kaggle website and is open source. This project is not intended to portray any gender or race in a positive or negative way, it's only attended to provide insights into the data itself.

#### Avg Wages by Year When Less Than High School Education

The scatter plot below shows that the year 1973 had the highest wage for people with less than a highschool education, paying on average a little over $18 hr. The year 1995 and 1996 had the lowest wage for people with less than a highschool education, only paying on average close to $14 hr.

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(x=wages_by_edu["less_than_hs"],y=wages_by_edu["year"], c="red", marker="+"),
plt.title("Wages by Year - Less Than High School Education", fontsize=20),
ticks = list(range(1970, 2026, 5))
plt.yticks(ticks, [f"{x}" for x in ticks]),
aph = list(range(14,19,1))
plt.xticks(aph, [f"${x} per hour" for x in aph], color="g"),
sns.set_style("darkgrid"),
style.use("ggplot")
```
