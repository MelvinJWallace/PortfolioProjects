1) Find the max salary earned by an employee.

###### *Utilizing the 'MAX' function to retrieve the value*

```python
max_salary = pd.read_sql("select max(salary) as max_salary from employee", con)
max_salary
```

2) Find the max salary earned by an employee in each department.

###### *Utilizing the 'MAX' function along with 'GROUP BY' to retrieve values*

```python
max_salary_by_dept = pd.read_sql("""select max(salary) as max_salary_per_dept, dept_name
                                 from employee
                                 group by dept_name""", con)
max_salary_by_dept
```

###### *Recreating the above value using python syntax and building a graph to show visualizations*

```python
max_salary_by_dept_visual = employee.groupby("DEPT_NAME")["SALARY"].max().reset_index().sort_values(by="SALARY", ascending=False)
sns.barplot(data=max_salary_by_dept_visual, x="SALARY", y="DEPT_NAME", palette="gist_stern") 
plt.title("Max Salary by Department",
          color="black",
          fontsize=15)
plt.ylabel("Department Name") 
plt.xlabel("") 
sal = list(range(0,13000,2000))
plt.xticks(sal, [f"${s}" for s in sal], color="green") 
print("The IT department has the highest salary earned by an employee with $11000.")
```

3) Find the max salary earned by an employee and show all rows.

###### *Utilizing the 'MAX' function along with 'GROUP BY' to retrieve the desired values*

```python
max_salary_of_employees = pd.read_sql("""select max(salary) as max_salary_of_employees, emp_name
                                      from employee
                                      group by emp_name
                                      order by max_salary_of_employees desc""", con)
max_salary_of_employees
```

4) Find the max salary earned by an employee in each department and show all cols.

###### *Utilizing a partition to group the data by 'DEPT_NAME' and the 'MAX' function to retrieve the desired values in the 'max_sal_by_dept' column*

```python
max_salary_by_dept_all_rows = pd.read_sql("""select *, max(salary) over(partition by dept_name order by salary desc) as max_sal_by_dept
                                          from employee""", con)
max_salary_by_dept_all_rows
```

5) Find the top employee by salary in each departments.

###### *Utilizing a partition to break the data up by 'DEPT_NAME', and the 'RANK' windows function with a common table expression to retrieve the values in the 'top_salary_by_dept' column* 

```python
top_emp_by_dept = pd.read_sql("""with top_emp as (select *, rank() over(partition by dept_name order by salary desc) as top_salary_by_dept
                                                  from employee)
                                 select top_emp.*
                                 from top_emp
                                 where top_salary_by_dept = 1
                                 order by salary desc""", con)
top_emp_by_dept
```

```python
sns.barplot(data=top_emp_by_dept, 
            x=top_emp_by_dept["SALARY"],
            y=top_emp_by_dept["emp_NAME"],
            hue=top_emp_by_dept["DEPT_NAME"],
            palette="gist_ncar") 
plt.title("Top Employee Salary by Department",
          color="black",
          fontsize=15)
plt.ylabel("Employee Name") 
plt.xlabel("") 
sal = list(range(0,13000,2000))
plt.xticks(sal, [f"${s}" for s in sal],
           color="green") 
print("The below graph displays the employees with the top salaries by department.") 
```

6) Find the first 2 employees who joined the company in each department.

###### *Utilizing a partition and the windows function 'ROW_NUMBER' in a subquery to retrieve the values in the 'first_two_emps' column*

```python
first_2_emps = pd.read_sql("""select *
                              from (select *, row_number() over(partition by dept_name order by emp_id asc) as first_two_emps
                                    from employee) f
                              where f.first_two_emps <= 2""", con)
first_2_emps
```

7) Find the top 3 employees in each department earning the max salary.

###### *Utilizing a partition and the windows function 'DENSE_RANK' in a common table expression to retrieve the values in the 'top_3_sal_by_dept' column*

```python
top_3_emps_by_dept = pd.read_sql("""with top_3 as (select *, dense_rank() over(partition by dept_name order by salary desc) as top_3_sal_by_dept
                                                   from employee)
                                    select top_3.*
                                    from top_3
                                    where top_3_sal_by_dept <= 3""", con)
top_3_emps_by_dept
```

8) Fetch a query to display if the salary of an employee is higher, lower, or equal to the previous salary of an employee in the same department.

###### *For this problem, I will demonstrate two ways of how this question can be answered. I will utilize a partition to break the data up by 'DEPT_NAME', and the windows functions 'LAG' and 'LEAD' to fetch values of the 'previous_salary' and 'next_salary' columns*

```python
emp_salary_1 = pd.read_sql("""select *, lag(salary) over(partition by dept_name order by emp_id asc) as previous_salary,
                                        lead(salary) over(partition by dept_name order by emp_id asc) as next_salary
                            from employee""", con)
emp_salary_1
```

```python
emp_salary_2 = pd.read_sql("""select *,
                                  case
                                      when salary < lag(salary) over(partition by dept_name order by emp_id asc) then 'less than previous employee salary'
                                      when salary = lag(salary) over(partition by dept_name order by emp_id asc) then 'equal to previous employee salary'
                                      when salary > lag(salary) over(partition by dept_name order by emp_id asc) then 'greater than previous employee salary'
                                      else 'first in department'
                                  end as previous_employee_salary
                              from employee""", con)
emp_salary_2
```

12) Write a query to segregate all the expensive phones, mid range phones and the cheaper phones.

###### *Utilizing the over clause and window function 'NTILE' with the case statement to retrieve the values in the 'phone price ranges' column*

```python
phone_seg = pd.read_sql("""select *,
                               case
                                   when phones = 1 then 'expensive phones'
                                   when phones = 2 then 'mid range phones'
                                   when phones = 3 then 'cheaper phones'
                               end as "phone price ranges"
                           from (select *, ntile(3) over(order by price desc) "phones"
                                 from product
                                 where product_category = 'Phone') p""", con)
phone_seg
```

13) Query to fetch all products which are constituting the first 30% of the data in products table based on price.

###### *Utilizing the 'ROUND' function, over clause and the window function 'CUME_DIST' along with a common table expression to retrieve the values in the 'cume' and 'percent' columns*

```python
first_30_percent = pd.read_sql("""with first_30 as (select *, cume_dist() over(order by price desc) "cume",
                                                              round(cume_dist() over(order by price desc), 2) * 100 "percent"
                                                    from product)
                                  select first_30.*
                                  from first_30
                                  where [percent] <= 30""", con)
first_30_percent
```

14) Query to identify how much percentage more expensive is "Galaxy Z Fold 3" when compared to all products.

###### *Utilizing the 'OVER' clause, the 'ROUND' function, and the window function 'PERCENT_RANK' with in a subquery to retrieve the values in the 'rank' and 'percent' columns*

```python
percent_more = pd.read_sql("""select *
                              from (select *, percent_rank() over(order by price asc) "rank",
                                              round(percent_rank() over(order by price asc), 2) * 100 "percent"
                                    from product) p
                              where p.product_name = 'Galaxy Z Fold 3'""", con)
percent_more
```

21) From the login details table, fetch the users who logged in consecutively 3 or more times.

###### *Utilizing the "CASE' statement with the windows function 'LEAD' in a subquery to retrieve the desired values*

```python
consecutive_logins = pd.read_sql("""select distinct c."consecutive user logins" 
                                    from (select *,
                                              case
                                                  when user_name = lead(user_name, 1) over(order by login_date asc)
                                                  and user_name = lead(user_name, 2) over(order by login_date asc)
                                                  then user_name
                                                  else null
                                              end as "consecutive user logins"
                                          from login_details) c
                                    where c."consecutive user logins" != 'None'""", con)
consecutive_logins
```

22) From the students table, write a SQL query to interchange the adjacent student names. 

###### *Utilizing the 'CASE' statement and the windows function 'LEAD' and 'LAG' to retrieve the values in the column 'adjacent student names'

```python
adjacent_student_names = pd.read_sql("""select *,
                                            case
                                                when id % 2 = 0 then lead(student_name, 1) over(order by id)
                                                when id % 2 != 0 then lag(student_name, 1, student_name) over(order by id)
                                            end as "adjacent student names"
                                        from students""", con)
adjacent_student_names
```

23) From the weather table, fetch all the records when London had extremely cold temperature for 3 consecutive days or more.

```python
consecutive_cold_days = pd.read_sql("""with t1 as (select *, row_number() over(order by id) "flag",
                                                             id - row_number() over(order by id) "diff"
                                                   from weather
                                                   where temperature <= 0),
                                            t2 as (select *, count(*) over(partition by t1."diff" order by t1."diff") "consecutive cold days"
                                                   from t1)
                                       select *
                                       from t2
                                       where "consecutive cold days" >= 3""", con)
consecutive_cold_days
```
